## Part 3: Data cleansing (missing values, outliers, duplicates, data consistently)

### 1. Filling missing value for the numerical columns.

```python
# Count the frequency of all elements in the columns "Retweets#"
data["Retweets#"].value_counts()
```

```python
## Remplace missing values in the columns "Retweets#" by 0 because 0 is the most frequency (2862 times)

data["Retweets#"].fillna((0), inplace=True)

## Remplace missing values by mean() for the other numeric columns

data["SenderFollowers#"].fillna(value=data["SenderFollowers#"].mean(), inplace=True)
data["AvgWordLength"].fillna(value=data["AvgWordLength"].mean(), inplace=True)
data["Emojis#"].fillna(value=data["Emojis#"].mean(), inplace=True)
data["AvgWordLength"].fillna(value=data["AvgWordLength"].mean(), inplace=True)
```

### 2. Normalize the type of the columns

```python
#Changing type of the data of the IsRetweet & IsSelfMentioned columns to booleans :

data['IsRetweet'] = data['IsRetweet'].astype('bool')
data['IsSelfMentioned'] = data['IsSelfMentioned'].astype('bool')
```

```python
#Changing the columns with the type of data float to int :

data['Id'] = data['Id'].astype('int64')
data['Retweets#'] = data['Retweets#'].astype('int64')
data['SenderId'] = data['SenderId'].astype('int64')
data['SenderFollowers#'] = data['SenderFollowers#'].astype('int64')
data['Emojis#'] = data['Emojis#'].astype('int64')
data['AvgWordLength'] = data['AvgWordLength'].astype('int64')
```

### 3. Finding and Removing Duplicates

The first thing we will do is attempt to drop any duplicate records, considering all the columns we currently have in the data set. Pandas provides us with the ability to do that via the drop_duplicates method. We will use the len method to calculate the number of rows in the data set both before and after removing duplicates and then print the number of rows dropped.

```python
before = len(data)
data = data.drop_duplicates()
after = len(data)
print('Number of duplicate records dropped: ', str(before - after))
```

### 4. Removing low variance column

We will want to strip away any columns that don't have a lot of value to us. One easy way to do this is to identify columns that have low variance, where the majority of the values in the column are the same. Since there is not a lot of variability in these columns, they have the potential to not be as informative as columns that have a variety of different values in them.

```python
## Find the columns with low variance

low_variance = []

for col in data._get_numeric_data():
    minimum = min(data[col])
    ninety_perc = np.percentile(data[col], 90)
    if ninety_perc == minimum:
        low_variance.append(col)

print(low_variance)
```

```python
## drop the low varaiance columns
data = data.drop(low_variance, axis=1)
```

### 5. Removing the unless column



```python
## find all columns of the data

data.columns
```

```python
## Removing the unless columns

unless_column=['Text','SenderAccountYears','Emojis#', 'Punctuations#',
       'UpperCaseLetter#', 'TWords#',
       'UWords#', 'SlangWords#','SenderLocation']
data = data.drop(unless_column, axis=1)
```

### 6. Extreme Values and Outliers

We should look for outliers, or extreme values, in the columns that remain. These outliers can influence our aggregations when we are analyzing data later, so we want to make sure we address them during our data cleaning stage.

```python
def find_outliers_by_decile(data, columns):
    outliers = pd.DataFrame(columns=data.columns)
    
    for col in columns:
        lower_decile = data[col].quantile(0.1)
        upper_decile = data[col].quantile(0.9)
        results = data[(data[col] < lower_decile) | (data[col] > upper_decile)].copy()
        results['Outlier'] = col
        outliers = outliers.append(results)
        
    return outliers

## Find the outlier for only 2 columns 'Retweets#','Favorites#'

outliers = find_outliers_by_decile(data, columns=['Retweets#','Favorites#'])
```

```python
outliers
```

"find_outliers_by_decile" is considering data points that are outside of the 10th and 90th percentile of each column as outliers. In this case, the function is only looking at the columns Retweets# and Favorites#, so the number of outliers is the total number of data points in these columns that fall outside of the 10th and 90th percentile.
The result of outlier have 999 rows, it's about 20% of the data. So in this case, we don't drop the outliers.
